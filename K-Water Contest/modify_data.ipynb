```python
import pandas as pd
```

```python
df_2015 = pd.read_excel('./data/정수장 일별 공급량/보령정수장/보령정수장_2015.xlsx')

# 데이터 재구성
melted_df_2015 = df_2015.melt(id_vars='구분(일)', var_name='월', value_name='데이터값')

# '월' 컬럼에서 숫자만 추출
melted_df_2015['월'] = melted_df_2015['월'].str.extract('(\d+)').astype(int)  # '1월'에서 '1'만 추출

# 'Date' 컬럼을 시도적으로 생성
try:
    melted_df_2015['Date'] = pd.to_datetime('2015-' + melted_df_2015['월'].astype(str) + '-' + melted_df_2015['구분(일)'].astype(str), errors='coerce')
except Exception as e:
    print(e)

# 잘못된 날짜 값을 갖는 행의 데이터값을 '-'로 설정
melted_df_2015.loc[melted_df_2015['Date'].isna(), '데이터값'] = '-'

# 잘못된 날짜 값을 갖는 행 제거
melted_df_2015 = melted_df_2015.dropna(subset=['Date'])

# '구분(일)'과 '월' 컬럼 제거
melted_df_2015 = melted_df_2015.drop(columns=['구분(일)', '월'])

# 원하는 순서대로 정렬
melted_df_2015 = melted_df_2015.sort_values(by='Date').reset_index(drop=True)

# '데이터값', 'Date' 순서로 컬럼 재정렬
melted_df_2015 = melted_df_2015[['Date', '데이터값']]

melted_df_2015
```
